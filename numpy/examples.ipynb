```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,10000)
x
```

```python
def f(x):
    return np.exp(-x/10)*np.sin(x)

y = f(x)
y
```

```python
plt.plot(x,y, c = 'red')
```

```python
x_prime = x[(x >= 4)*(x <= 7)]
x_prime
```

```python
y_prime = f(x_prime)
y_prime
```

```python
np.mean(y_prime)
```

```python
np.std(y_prime)
```

```python
y_m = np.percentile(y_prime,80)
y_m
```

```python
dydx = np.gradient(y,x)
plt.plot(dydx,x,c='b')
```
