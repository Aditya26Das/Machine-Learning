```python
def f(x):
    return x**2 * np.sin(x)/np.exp(-x)

x = np.linspace(0,10,100)
y = f(x)
plt.plot(x,y)
plt.show()
```

#### Calculus / Statistics

```python
n1 = np.random.randn(100)
## A normal distribution is generated with mean zero and median 1
n1
```

```python
n2 = 2*n1
## Standard distribution is doubled while mean is zero
n2
```

```python
n3 = n2 + 10
## mean is increased by 10
n3
```

```python
np.percentile(n1,80)
```

#### Integrals and Derivatives

```python
x = np.linspace(1,10,100)
y = 1/x**2 * np.sin(x)
dydx = np.gradient(y,x)
y_in = np.cumsum(y) * (x[1] - x[0])
plt.plot(x,dydx , c= 'r')
plt.plot(x,y_in, c = 'y')
plt.plot(x,y, c = 'b')
```
