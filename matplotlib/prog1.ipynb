```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

### 2D line plot

Used for bivariate analysis
-->Numerical-Numerical
-->Numerical-Categorical

```python
price = []
for i in range(1,11):
    price.append(np.random.randint(1000,10000))

year = []
for i in range(2010,2020):
    year.append(i)

plt.plot(price,year)
```

```python
batsman = pd.read_csv('../matplotlib/datasets/sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
plt.plot(batsman['index'],batsman['RG Sharma'],c='b',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['V Kohli'],c='r',linestyle='dashdot',linewidth=2)
plt.title('Carrier Comparison of Rohit Sharma Vs Virat Kohli')
plt.xlabel('Season of IPL--->')
plt.ylabel('Runs Scored--->')
plt.legend(['Sharma','Kohli'])
plt.grid()
```

### Bar Chart

-->used for bivariate analysis in categorical-numerical data 

```python
df = pd.read_csv('../matplotlib/datasets/batsman_season_record.csv')
df
```

```python
new_df = df.drop(['2015','2016'],axis = 1)
new_df
```

```python
plt.bar(new_df['batsman'],new_df['2017'])
```

```python
plt.barh(new_df['batsman'],new_df['2017'])
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

### QQ Plot

```python
import statsmodels.api as sm 
import matplotlib.pyplot as plt 
import numpy as np

x = np.random.normal(5,6,1000)
sm.qqplot(x,line='45',fit=True)

plt.show()
```
