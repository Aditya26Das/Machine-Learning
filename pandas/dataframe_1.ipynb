### Creating Dataframes

```python
import pandas as pd
import numpy as np
```

#### using lists

```python
student_marks = [
    [100,100,20],
    [90,80,15],
    [70,90,17]
]

pd.DataFrame(student_marks,columns=['Marks','IQ','Package'])
```

#### using dictionary

```python
student_marks = {
    'Marks': [100,90,70],
    'IQ': [100,80,90],
    'Package': [20,15,17]
}

pd.DataFrame(student_marks)
```
